```python
import pandas as pd
from transformers import pipeline, AutoTokenizer
```

```python
# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained('w11wo/indonesian-roberta-base-sentiment-classifier')
```

```python
df = pd.read_csv('cleaned_add_merged_data.csv')
```

```python
# Initialize the sentiment analysis pipeline
classifier = pipeline(
    'text-classification',
    model='w11wo/indonesian-roberta-base-sentiment-classifier',
    tokenizer=tokenizer,
    framework='pt',  # Specify the framework (PyTorch)
    max_length=512,  # Set the maximum sequence length
    truncation=True  # Enable truncation if necessary
)
```

```python
# Function to label sentiment for a given text
def label_sentiment(text):
    result = classifier(text)
    label = result[0]['label']
    confidence = result[0]['score']
    return label, confidence
```

```python
# Modify the labeling function to map 'neutral' to 'positive'
def label_and_modify_sentiment(text):
    label, confidence = label_sentiment(text)
    if label == 'neutral':
        label = 'positive'
    return label, confidence
```

```python
# Apply the modified labeling function to the 'text' column and create new columns 'Sentiment' and 'Confidence'
df[['Sentiment', 'Confidence']] = df['Review'].apply(label_and_modify_sentiment).apply(pd.Series)
```

```python
df.to_csv('labeled_cleaned_add_tweets_merged_data.csv', index=False)
```
