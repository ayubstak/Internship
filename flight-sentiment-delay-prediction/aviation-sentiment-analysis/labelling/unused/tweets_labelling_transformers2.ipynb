```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('cleaned_tweets_super air jet.csv')
```

```python
from googletrans.client import Translator
translator = Translator()

from transformers import pipeline
sentiment_classifier = pipeline('sentiment-analysis', framework='pt', max_length=512, truncation=True)
```

```python
def clean_tweet(tweet):
  return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\s+)", " ", tweet).split())
```

```python
data['label'] = data['content'].str.encode('ascii','ignore').apply(translator.translate, dest='en')
data['label'] = data['label'].apply(getattr, args=('text',))
data['label'] = data.apply(lambda x: clean_tweet(x['label']), axis=1)
```

```python
df = (
    data
    .assign(sentiment = lambda x: x['label'].apply(lambda s: sentiment_classifier(s)))
    .assign(
        label = lambda x: x['sentiment'].apply(lambda s: (s[0]['label'])),
        score = lambda x: x['sentiment'].apply(lambda s: (s[0]['score']))
    )
)
```

```python
df.sample(20)
```

```python
df.to_csv('labeled_transformer2_tweets_super air jet.csv', index=False)
```
